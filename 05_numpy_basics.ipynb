## Learning goals
- Create and manipulate arrays
- Perform vectorized math
- Understand axis-based operations

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:", matrix)
print("Column sums:", matrix.sum(axis=0))
print("Row sums:", matrix.sum(axis=1))
```

```python
# Practice task:
# Generate 100 random numbers and compute min, max, mean.
nums = np.random.rand(100)
print("Min:", nums.min())
print("Max:", nums.max())
print("Mean:", nums.mean())
```
