```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

sales = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "revenue": [1200, 1350, 1280, 1500, 1650, 1720],
})

plt.figure(figsize=(8, 4))
plt.plot(sales["month"], sales["revenue"], marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
students = pd.DataFrame({
    "group": ["A", "A", "B", "B", "C", "C"],
    "score": [70, 75, 80, 85, 90, 95]
})

plt.figure(figsize=(6, 4))
sns.boxplot(data=students, x="group", y="score")
plt.title("Score Distribution by Group")
plt.show()
```

```python
# Practice task:
# Create a histogram of the score column and explain what it shows.
plt.figure(figsize=(6, 4))
sns.histplot(students['score'], bins=6, kde=False)
plt.title('Histogram of Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()

# Explanation
print('Explanation: The histogram shows the distribution of student scores.')
print('In this sample the scores range from 70 to 95 and each score value is represented once,')
print('so the histogram shows one count per score bin and no single concentration;')
print('with larger datasets, the histogram reveals clustering, spread, and skew.')
```
