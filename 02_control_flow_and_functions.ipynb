# 02 — Control Flow and Functions

## Learning goals
- Use if/elif/else for decisions
- Use loops for repetition
- Write reusable functions

```python
score = 83

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

```python
# Loops
nums = [2, 4, 6, 8]
total = 0
for n in nums:
    total += n

print("Total:", total)
```

```python
# Functions
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(25))
```
