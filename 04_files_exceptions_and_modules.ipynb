```python
# Exception handling
raw = ["10", "20", "oops", "30"]
parsed = []

for item in raw:
    try:
        parsed.append(int(item))
    except ValueError:
        print(f"Skipping non-numeric value: {item}")

print("Parsed:", parsed)
```

```python
# Practice task:
# Create a function safe_divide(a, b) that handles division by zero.
def safe_divide(a, b):
    """Return the division a / b, or None if b is zero."""
    try:
        return a / b
    except ZeroDivisionError:
        print("Cannot divide by zero.")
        return None

# Examples
print(safe_divide(10, 2))  # 5.0
print(safe_divide(5, 0))   # None (handled)
```
