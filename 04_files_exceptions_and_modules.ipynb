```python
# Exception handling
raw = ["10", "20", "oops", "30"]
parsed = []

for item in raw:
    try:
        parsed.append(int(item))
    except ValueError:
        print(f"Skipping non-numeric value: {item}")

print("Parsed:", parsed)
```
